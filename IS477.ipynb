```python
import os, requests
from pprint import pprint

email = os.getenv("AQS_EMAIL")
key = os.getenv("AQS_KEY")

print("AQS_EMAIL:", email)
print("AQS_KEY:", key[:4] + "..." if key else None)

base_url = "https://aqs.epa.gov/data/api/dailyData/byState"

params = {
    "email": email,
    "key": key,
    "param": "88101",    
    "state": "17",       
    "bdate": "20240101",
    "edate": "20240101",
}

response = requests.get(base_url, params=params)
print("Status code:", response.status_code)

data = response.json()
print("Header:")
pprint(data.get("Header"))
print("Number of records:", len(data.get("Data", [])))
```

```python
import pandas as pd 

records = data.get("Data", [])
df = pd.DataFrame(records)

print("Shape:", df.shape)
df.head()
```
